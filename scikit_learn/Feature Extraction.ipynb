```
from sklearn.feature_extraction.text import CountVectorizer
```

```
messages = ["Hello, my name is Rudra",
            "I'm a Machine Learning Engineer",
            "I'm a big fan of Elon Musk",
            "Chess and Stock Market are my pass-time hobbies",
            "I love eating cheese burgers"]
```

```
# Using CountVectorizer, we'll be treating each unique word as a feature
vect = CountVectorizer()
x = vect.fit_transform(messages)
print(vect.get_feature_names())
len(vect.get_feature_names())
```

```
%%writefile random.txt
Hello, Machine Learning enthusiasts!
This code is written by Rudra.
He is 25 years young.
He is an incoming Data Scientist in the world.
```

```
vocab = {}
i = 1
y = ' '.join(messages).split()

# assigning a key to each word
for word in y:
  if word in vocab:
    continue
  else:
    vocab[word] = i
    i+=1
print(vocab)
```

```
with open('random.txt') as f:
  x = f.read().lower().split() ### splitting the words of the file in a dictionary

# assigning a key to each word
for word in x:
  if word in vocab:
    continue
  else:
    vocab[word] = i
    i+=1
print(vocab)
```

```
# Create an empty vector with space for each word in the vocabulary
one = ['random.txt']+[0]*len(vocab)
one
```

```
# Map the frequency of each word in random.txt to our vector
with open('random.txt') as f:
  x = f.read().lower().split() ### splitting the words of the file in a dictionary

for word in x:
  one[vocab[word]] += 1 # count the freq

one
```
